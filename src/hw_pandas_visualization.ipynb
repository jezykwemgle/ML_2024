Зріст: піковий зріст це ~ 210 см

Вага: пікова вага це ~ 100 кг
Загалом найрозповсюдженіша вага гравців знаходиться в проміжку від 70 до 130 кг

```python
players['BMI'] = players['weight'] / ((players['height'] * 0.01) ** 2)
players['BMI'].head(10)
```

```python
players['BMI'].plot(kind="density", subplots=True, layout=(1, 1), sharex=False, figsize=(10, 7))
```

На графіку видно помітний пік навколо ІМТ 25. Це вказує на те, що найпоширеніше значення ІМТ серед людей з набору даних близьке до 25, що, згідно з класифікацією ІМТ, знаходиться на межі між "нормальною" та "надмірною" вагою.
Графік не демонструє значного перекосу ні вліво, ні вправо, а це означає, що немає екстремальних значень, які б витягали розподіл занадто далеко в будь-який бік тому можна вважати що розподіл нормальний.

```python
import numpy as np

conditions = [
    (players['BMI'] <= 18.4),
    (players['BMI'] > 18.4) & (players['BMI'] <= 24.9),
    (players['BMI'] > 24.9) & (players['BMI'] <= 39.9),
    (players['BMI'] >= 40.0)
]
choices  = ['Underweight', 'Normal', 'Overweight', 'Obese']
players['Status'] = np.select(conditions, choices, default='Unknown')
players[['height', 'weight', 'BMI', 'Status']].head(10)
```

```python
import featuretools as ft

es = ft.EntitySet(id='players_data')

es = es.add_dataframe(dataframe_name='players',
                      dataframe=players[['height', 'weight', 'born']],
                      index='Unnamed')

feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_dataframe_name='players',
                                      trans_primitives=['add_numeric', 'multiply_numeric', 'subtract_numeric', 'divide_numeric'])

feature_matrix.head()
```

```python
features = ["PER", "WS", "DWS", "OWS", "WS/48", "BPM"]
merged_data[features].plot(kind="density", subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10))
```

Графіки показують, що більшість гравців має низькі до середніх показники за кожною метрикою з невеликою кількістю гравців, що виділяються високими або дуже низькими значеннями.

```python
merged_data['year_start'] = pd.to_datetime(merged_data['year_start'].astype(str), format='%Y')
merged_data['birth_date'] = pd.to_datetime(merged_data['birth_date'])

# Розрахунок віку гравців на момент початку їх кар'єри в новій колонці
merged_data['age_start'] = merged_data['year_start'] - merged_data['birth_date']
merged_data['age_start'] = (merged_data['age_start'] / pd.Timedelta(days=365)).astype(float)

sns.histplot(merged_data['age_start'], bins=20, kde=True)
plt.title('Віковий розподіл гравців')
plt.xlabel('Age at start of career')
plt.ylabel('Count')
plt.show()
```

Більшість гравців починають свою кар'єру в віці між 22 і 25 роками. Найвищий пік припадає на вік близько 23 років
Графік має форму дзвону, що свідчить про нормальний розподіл віку гравців на момент початку кар'єри.
Хоча основна маса гравців починає кар'єру в зазначеному віці 22-25 років, існують також гравці, що починають її значно раніше або пізніше.

```python
# Player Efficiency Rating (PER) Over Years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='PER', data=merged_data)
plt.title('Рейтинг ефективності гравців за роками')
plt.xlabel('Year')
plt.ylabel('PER')
plt.show()
```

Показник ефективності (PER) з часом зростає. Починаючи з 1950-х років і до кінця 1960-х, існує тренд на зниження, після чого починається стабілізація та поступовий ріст.
Рейтинг ефективності має певні коливання, особливо в період з 1970 до 1980 років
Після 1980-х років графік показує більш стабільний і послідовний ріст, з деякими коливаннями, але загальний тренд залишається позитивним.
